# Data Cleaning in Pandas

In this Colab notebook, we will walk through the process of cleaning and preprocessing a dataset using the Python Pandas library. Data cleaning is a crucial step in the data analysis and data science pipeline, as it ensures that the data you work with is accurate, consistent, and ready for analysis.

The dataset used in this notebook is called "Customer Call List.xlsx." We will perform several data cleaning operations to improve the quality and usability of the data.

Here's a summary of the main data cleaning steps performed in this notebook:

1. **Loading the Data:** We start by loading the dataset into a Pandas DataFrame using the `pd.read_excel` function.

2. **Removing Duplicate Rows:** Duplicate rows can skew analysis results, so we use the `drop_duplicates` method to remove them.

3. **Dropping Unnecessary Columns:** We drop a column named "Not_Useful_Column" that is not needed for our analysis.

4. **Cleaning Text Data:** We clean the "Last_Name" column by removing specific characters like digits, underscores, dots, and slashes. Similarly, we clean the "Phone_Number" column by replacing certain patterns with empty strings.

5. **Splitting Address Information:** We split the "Address" column into separate columns for "Street_Address," "State," and "Zip_Code" to make the data more structured.

6. **Standardizing "Do_Not_Contact" Values:** We standardize the values in the "Do_Not_Contact" column, replacing "Yes" with "Y" and "No" with "N."

7. **Handling Missing Values:** We fill any remaining missing values in the dataset with empty strings.

8. **Filtering Rows:** We remove rows where the "Do_Not_Contact" column is marked as "Y" and rows where the "Phone_Number" is empty, as these are not relevant for our analysis.

9. **Resetting the Index:** Finally, we reset the DataFrame index to ensure it is sequential and starts from zero.

This notebook serves as a practical example of common data cleaning techniques that are essential for preparing data for further analysis or machine learning tasks. By the end of this notebook, the dataset will be in a more suitable state for various analytical tasks.


```python
# Import the pandas library
import pandas as pd
```

```python
# Read the data from an Excel file into a pandas DataFrame
df = pd.read_excel(r"Customer Call List.xlsx")
df
```

```python
# Remove duplicate rows from the DataFrame
df = df.drop_duplicates()
df
```

```python
# Drop a column named "Not_Useful_Column"
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Remove leading and trailing characters from the "Last_Name" column
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# Remove 'nan--' from the "Phone_Number" column
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

# Remove 'Na--' from the "Phone_Number" column
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# Split the "Address" column into "Street_Address," "State," and "Zip_Code" columns
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```python
# Replace 'Yes' with 'Y' and 'No' with 'N' in the "Do_Not_Contact" column
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
# Fill NaN (missing) values with empty strings
df=df.fillna('')
df
```

```python
# Iterate through the DataFrame to remove rows where "Do_Not_Contact" is 'Y'
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
# Iterate through the DataFrame to remove rows where "Phone_Number" is empty
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```

```python
# Reset the index of the DataFrame
df = df.reset_index(drop=True)
df
```
